### Pre processing Data

```python
# To check any null values present in the data.
df.isnull().sum()
```

```python
# The unwanted data to be deleted using drop fucntion.
data = df.drop("Unnamed: 0", axis = 1)
data
```

```python
# To find out the unique values from the city.
data["JobCity"].unique()
```

```python
# To convert into lower aplhabet for distribution
data["JobCity"].str.lower()
```

```python
# Count -1 values across all columns
count_negative_ones_all = (data == -1).sum()
print(count_negative_ones_all)
```

```python
# To convert present value into present date for good distribution and the convert into datetime format.
data["DOL"].replace("present",pd.Timestamp.today(),inplace=True)
data['DOL'] = pd.to_datetime(data['DOL']).dt.date
data['Period'] = pd.to_datetime(data["DOL"]).dt.year - pd.to_datetime(data['DOJ']).dt.year
data
```

```python
# To get the data which has designation = get
data[data["Designation"]=="get"][['Designation','JobCity','Salary','Specialization']]
```

```python
# merge the specialization into a variable so we can ascess easily.
mech = data[data['Specialization'].isin(['mechanical engineering','mechanical and automation'])]['Designation'].mode()[0]
eee = data[data['Specialization']==('electronics and electrical engineering')]['Designation'].mode()[0]
print(f'mode for mechanical:  {mech}\nmode for EEE:  {eee}')
```

```python
# To replcae the get from the variable that we created.
data.loc[data['Specialization'].isin(['mechanical engineering','mechanical and automation']),'Designation'].replace('get',mech,inplace=True)

data['Designation'].replace('get',eee,inplace=True)
```

```python
# To convert -1 values in the data into unknown
data['JobCity'].replace(-1,'unknown',inplace=True)
data['JobCity'].apply(lambda x:x.title())
```

```python
# To find the unknown in the data.
data[data["JobCity"]=='unknown']
```

```python
# To correct the spelling of some cities.
data["JobCity"].replace("Bangalore","Bengaluru",inplace=True)
data["JobCity"].replace("Banaglore","Bengaluru",inplace=True)
data["JobCity"].replace("Chennai, Bangalore","Bengaluru",inplace=True)
data["JobCity"].replace(" Bangalore","Bengaluru",inplace=True)
data["JobCity"].replace("Bangalore ","Bengaluru",inplace=True)
data["JobCity"].replace("Banglore","Bengaluru",inplace=True)
data["JobCity"].replace("Jaipur ","Jaipur",inplace=True)
data["JobCity"].replace("Gandhinagar","Gandhi Nagar",inplace=True)
data["JobCity"].replace("Bangalore ","Bengaluru",inplace=True)
data["JobCity"].replace("Jaipur ","Jaipur",inplace=True)
data["JobCity"].replace("Gandhinagar","Gandhi Nagar",inplace=True)
data["JobCity"].replace("Hyderabad ","Hyderabad",inplace=True)
data["JobCity"].replace("Hyderabad(Bhadurpally)","Hyderabad",inplace=True)
data["JobCity"].replace("Bhubaneswar ","Bhubaneswar",inplace=True)
data["JobCity"].replace("Delhi/Ncr","Delhi",inplace=True)
data["JobCity"].replace("Nagpur ","Nagpur",inplace=True)
data["JobCity"].replace("Pune ","Pune",inplace=True)
data["JobCity"].replace("Trivandrum ","Trivandrum",inplace=True)
data["JobCity"].replace("Thiruvananthapuram","Trivandrum",inplace=True)
```
